```python
### TPC4: Aplicação para Gerir um Cinema

#Cinema = [Sala]
#Sala = (nlugares, Vendidos, filme)
#nlugares = Int
#filme = String 
#Vendidos = [Int]

def menu():
    print("""Bem-vindo ao cinema!
Menu:
(1) Filmes em exibição no cinema
(2) Procurar lugar disponível
(3) Vender bilhete
(4) Lista dos lugares ocupados
(5) Inserir uma nova sala
(0) Sair""")
    return

sala1 = ("sala1", 150, [], "Twilight")
sala2 = ("sala2", 200, [], "Hannibal")
cinema = [sala1,sala2]
filmes=("Twilight","Hannibal")

def listar(cinema):
    filmes=[]
    for sala in cinema:
        nlugares, vendidos , nomefilme = sala
        filmes.append(nomefilme)
    
    return filmes

def disponivel(cinema,nomefilme,lugar):
    condiçao=False
    for i in cinema:
        nlugares, vendidos , filme = i
        if nomefilme==filme and lugar<=nlugares:
            if lugar not in vendidos:
                condiçao=True
    return condiçao

def vendebilhete(cinema,nomefilme,lugar):
    for i in cinema:
        var=0
        nlugares,vendidos,filme=i
        if nomefilme==filme:
            if lugar in range(nlugares):
                if lugar not in vendidos:
                    vendidos.append(lugar)
                    vendidos.sort()
                else:
                    var=1
    return var

def listardisponibilidade(cinema):
    lista=[]
    for i in cinema:
        nlugares,vendidos,filme=i
        lista.append((filme,nlugares-len(vendidos)))
    return 



def inserirsala(cinema,nomesala,nlugares,nomefilme):
    salanova= (nomesala,nlugares, [], nomefilme)
    cinema.append(salanova)
    return cinema


menu()
cinema=[]
function=input("O que pretende fazer?")

while function!=0:
    if function==1:
        print("No cinema estão em exibição os seguintes filmes:",listar(filmes))
       
    if function==2:
        condiçao=disponivel(input("Que lugar pretende selecionar?"))
        if condiçao == False:
            print("")
            print("O lugar está ocupado!")
        if condiçao == True:
            print("")
            print("O lugar está livre!")

    if function==3:
        cinema,var=vendebilhete(cinema,input("Para que sala pretende ir?"),int(input("Em que lugar se pretende sentar?")))
        if var==1:
            print("")
            print("O lugar onde pretende sentar já está ocupado.")

    if function==4:
        nome,lugaresLivres=listardisponibilidade(cinema)
        print("")
        print("A sala",nome,"tem",lugaresLivres,"lugar(es) livre(s)")

    if function==5:
        nomesala=input("Qual o nome da sala?")
        nlugares= int(input("Quantos lugares vai ter a nova sala?"))
        nomefilme=input("Qual o nome do filme?")
        if nomesala in cinema:
            print("Já existe uma sala com esse nome.")
        else:
            inserirsala(cinema,nomesala,int(input("Quantos lugares terá a nova sala?")),input("Que filme será exibido na nova sala?"))
print("")

```
