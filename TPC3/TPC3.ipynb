```python
from random import random
from turtle import pos


def menu ():
    print(''' Menu:
(1) Criar Lista 
(2) Ler Lista
(3) Soma
(4) Média
(5) Maior
(6) Menor
(7) estaOrdenada por ordem crescente
(8) estaOrdenada por ordem decrescente
(9) Procura um elemento
(0) Sair''')
    return

import random
def criaListaAleatoria(N):
    n=0
    listaaleatoria=[]
    while n<=N:
        elemaleatorio= random.randint (1,101) 
        listaaleatoria.append(elemaleatorio)
        n=n+1
    print (listaaleatoria) 
    return listaaleatoria

def leLista(N):
    lista1= []
    n=0
    while N>0:
        n = int(input("insira um numero"))
        lista1.append(n)
        N=N-1
    return lista1


def somaLista(lista):
    soma=0
    for n in lista:
        soma=soma+n
    return soma

def mediaLista(listamedia):
    media=somaLista(listamedia)/len(listamedia)
    return media

def maiorLista(lista):
    maior=lista[0]
    for i in lista:
        if i>maior:
            maior=i
    return maior

def menorLista(lista):
    menor=lista[0]
    for i in lista:
        if i<menor:
            menor=i
    return menor

def ordemcrescente(lista):
    valor=True
    i=0
    while i+1<len(lista):
        if lista[i]>lista[i+1]:
            valor=False
        i=i+1    
    return valor

def ordemdecrescente(lista):
    valor=True
    i=0
    while i+1<len(lista):
        if lista[i]<lista[i+1]:
            valor=False
        i=i+1    
    return valor

def pos(lista,elem):
    procuraLista=[]
    for i in range(len(lista)):
        if lista[i]==elem:
            procuraLista.append(i+1)
    if len(procuraLista)==0:
        procuraLista.append(-1)
    return procuraLista

menu()
function=int(input("o que pretende selecionar?"))
listafinal=[]
while function!=0:
    if function==1:
        listafinal.clear()
        listafinal=criaListaAleatoria(int(input("qual o tamanho da lista?")))
    if function==2:
        listafinal.clear()
        listafinal=leLista(int(input("qual o tamanho da lista?")))
    if function==3:
        print("")
        print("a soma dos elementos é:",somaLista(listafinal))
    if function==4:
        print("")
        print("a média dos elementos é:",mediaLista(listafinal))
    if function==5:
        print("")
        print("o maior elemento é:",maiorLista(listafinal))
    if function==6:
        print("")
        print("o menor elemento é:",menorLista(listafinal))
    if function==7:
        if ordemcrescente(listafinal)== True:
            print("")
            print("sim")
        else:
            print("")
            print("não")
    if function==8:
        if ordemdecrescente(listafinal)==True:
            print("")
            print("sim")
        else:
            print("")
            print("não")
    if function==9:
        procuraLista=pos(listafinal,int(input("Qual o elemento que procura?")))
        if procuraLista[0]==-1:
            print("")
            print("O elemento que procura não se encontra na lista.")
        else:
            print("")
            print("o elemento encontra-se na ",procuraLista,"posição")
    function=int(input("o que pretende selecionar?"))
print("")
print("a lista é:",listafinal)
        
```
